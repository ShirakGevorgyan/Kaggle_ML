```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
drop_values = np.abs(df_train.corr()['yield']).sort_values()[:-4].index
```

```python
drop_values
```

```python
df_train = df_train.drop(drop_values, axis=1)
df_test = df_test.drop(drop_values, axis=1)
```

```python
X = df_train.drop('yield', axis=1)
y = df_train['yield']
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
params = {
    "criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter":["best", "random"],
    "max_depth": [2, 4, 8, None],
    "min_samples_split": [2, 3, 4, 5],
    "min_samples_leaf":[1, 2, 3, 4],
    "max_features": [ 'sqrt', 'log', None],
    "max_leaf_nodes" : [2, 3, 4, 5]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
final_model = GridSearchCV(model, param_grid=params, verbose=2, cv=2)
```

```python
final_model.fit(X, y)
```

```python
y_pred = final_model.predict(df_test)
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['yield'] = y_pred
```

```python
sub.to_csv('sub.csv', index=False, index_label=False)
```
