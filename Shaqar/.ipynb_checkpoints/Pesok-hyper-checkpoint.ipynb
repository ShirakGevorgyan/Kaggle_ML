```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
df_test = pd.read_csv('test_data.csv')
df_train = pd.read_csv('train.csv')
df_sub = pd.read_csv('submission_form.csv')
```

```python
df_test
```

```python
df_train['8'].value_counts()
```

```python
df_train.corr()['8']
```

```python
df_test = df_test.drop('Unnamed: 0', axis=1)
df_train = df_train.drop('Unnamed: 0' , axis=1)
```

```python
df_train.describe()
```

```python
X = df_train.drop('8', axis=1)
```

```python
y = df_train['8']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
mydict = []
for i in range(5,51):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X,y)
    mydict.append(model.score(X, y))
```

```python
best_n = mydict.index(max(mydict)) + 5
```

```python
best_n
```

```python
model = KNeighborsClassifier(n_neighbors=best_n)
```

```python
model.fit(X, y)
```

```python
df_test=df_test.drop('8', axis=1)
```

```python
y_pred = model.predict(df_test)
```

```python
y_pred
```

```python
kl = []
for i in y_pred:
    kl.append(int(i/1))
```

```python
kl
```

```python
df_sub['Label'] = kl
```

```python
df_sub.to_csv('pesok.csv', index=False, index_label=False)
```
