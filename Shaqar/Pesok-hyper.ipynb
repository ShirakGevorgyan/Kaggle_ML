```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
df_test = pd.read_csv('test_data.csv')
df_train = pd.read_csv('train.csv')
df_sub = pd.read_csv('submission_form.csv')
```

```python
df_test = df_test.drop('Unnamed: 0', axis=1)
df_train = df_train.drop('Unnamed: 0' , axis=1)
```

```python
X = df_train.drop('8', axis=1)
```

```python
y = df_train['8']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = KNeighborsClassifier()
```

```python
params = {
    'n_neighbors':[i for i in range(5,12,2)],
    'weights':['uniform', 'distance'],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size':[30, 40, 50],
    'p':[2, 4, 6],
    'metric':['minkowski', 'euclidean'],
    'n_jobs':[None, 2, 4]
}
```

```python
ful_model = GridSearchCV(model, params, cv = 5, verbose=2)
```

```python
ful_model.fit(X, y)
```

```python
# df_test=df_test.drop('8', axis=1)
```

```python
y_pred = ful_model.predict(df_test)
```

```python
y_pred
```

```python
kl = []
for i in y_pred:
    kl.append(int(i/1))
```

```python
df_sub['Label'] = kl
```

```python
df_sub.to_csv('pesok.csv', index=False, index_label=False)
```
