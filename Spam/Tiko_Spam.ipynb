```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in df_train.columns:
    if df_train[i].dtype == object:
        label = LabelEncoder()
        df_train[i] = label.fit_transform(df_train[i])
```

```python
for i in df_test.columns:
    if df_test[i].dtype == object:
        label = LabelEncoder()
        df_test[i] = label.fit_transform(df_test[i])
```

```python
df_train.corr()["Exited"]
```

```python
df_test = df_test.drop(["id" , "CustomerId", "Surname", "Tenure" , "HasCrCard" , "EstimatedSalary"] , axis =1 )
```

```python
df_train = df_train.drop(["id" , "CustomerId", "Surname", "Tenure" , "HasCrCard" , "EstimatedSalary"] , axis =1 )
```

```python
df_test.info()
```

```python
for i in df_train.columns:
    p_75 = float(df_train.describe()[i].loc['75%'])
    p_25 = float(df_train.describe()[i].loc['25%'])
    IQR = float(df_train.describe()[i].loc['75%']) - float(df_train.describe()[i].loc['25%'])
    df_train[i] = df_train[i].apply(lambda x : np.mean(df_train[i]) if x < p_25 - IQR * 1.5 or x > p_75 + IQR * 1.5 else x)
```

```python
for i in df_test.columns:
    p_75 = float(df_test.describe()[i].loc['75%'])
    p_25 = float(df_test.describe()[i].loc['25%'])
    IQR = float(df_test.describe()[i].loc['75%']) - float(df_test.describe()[i].loc['25%'])
    df_test[i] = df_test[i].apply(lambda x : np.mean(df_test[i]) if x < p_25 - IQR * 1.5 or x > p_75 + IQR * 1.5 else x)
```

```python
X = df_train.drop("Exited" ,axis = 1)
y = df_train["Exited"]
```

```python
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
hyper_param = {
    'n_estimators': [64, 100],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10],
    # 'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'min_weight_fraction_leaf': [0.0, 0.1],
    'max_leaf_nodes': [None, 10],
    'min_impurity_decrease': [0.0, 0.1],
#     'bootstrap': [True, False],
#     'oob_score': [False, True],
#     'n_jobs': [None, -1],
#     'random_state': [None, 42],
#     'verbose': [0, 1],
#     'warm_start': [False, True],
#     'class_weight': [None, 'balanced'],
#     'ccp_alpha': [0.0, 0.1],
#     'max_samples': [None, 0.5]
}
```

```python
full_model = GridSearchCV(model, hyper_param, cv=3, verbose=2)
```

```python
full_model.fit(X, y)
```

```python
y_pred = full_model.predict(df_test)
```
