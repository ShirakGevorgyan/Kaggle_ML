```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('train.csv')
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
df_train.corr()['FloodProbability']
```

```python
df_test = df_test.drop('id', axis=1)
df_train = df_train.drop('id', axis=1)
```

```python
X = df_train.drop('FloodProbability', axis=1)
y = df_train['FloodProbability']
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
sceler = StandardScaler()
min_max = MinMaxScaler()
```

```python
X = sceler.fit_transform(X)
```

```python
y.unique()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestRegressor()
```

```python
hyper_params = {
    'n_estimators': [2],
    'max_depth': [None],
    'min_samples_split': [2],
    'min_samples_leaf': [1]
}
```

```python
full_model = GridSearchCV(model, hyper_params, cv=2, verbose=2)
```

```python
full_model.fit(X, y)
```

```python
y_pred = full_model.predict(df_test)
```

```python
df_sub= pd.read_csv('sample_submission.csv')
```

```python
df_sub["FloodProbability"] = y_pred
```

```python
df_sub.to_csv('sub_trch.csv', index=False, index_label=False)
```
